confusion Matrix

Confusion metrices is a table which is used to describe the perfomance of a classification problem, it visualize the accuracy of a classifier by comparing the predicted values with actual values,
terms like TP-True Postive,TN- True Negative, FP- False postive,FN- False Negative

True Postive: The predicted result is Postive, While it is labeled as postive,
False Postive:The predicted result is Postive, While it is labeled as Negative, it calls type 1 error as well,
False Negative: The predicted result is Negative, While it is labeled as postive, it calls type 2 error as well,
True Negative:The predicted result is Negative, While it is labeled as Negative

confusion matrix of Logistic Regression:Logistic regression is a method to predict a binary outcome, such as yes or no, based on prior observations of a data set.


```python
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(y_test,lr_pred),annot=True,fmt="d")
cm_logreg = confusion_matrix(y_test,lr_pred)
```

```python
#defining function to calculate specificty and sensitivity
def calculate_specificity_sensitivity(y_true, y_pred):
    TN, FP, FN, TP = confusion_matrix(y_true, y_pred).ravel()
    specificity = TN / (TN + FP)
    sensitivity = TP / (TP + FN)
    return specificity, sensitivity
```

```python
# Calculate the specificity and sensitivity for each model
specificity_logreg, sensitivity_logreg = calculate_specificity_sensitivity(y_test, lr_pred)
specificity_knn, sensitivity_knn = calculate_specificity_sensitivity(y_test, knn_pred)
specificity_nb, sensitivity_nb = calculate_specificity_sensitivity(y_test, nb_pred)
specificity_svm, sensitivity_svm = calculate_specificity_sensitivity(y_test, sv_pred)
specificity_dt, sensitivity_dt = calculate_specificity_sensitivity(y_test, dt_pred)
specificity_rf, sensitivity_rf = calculate_specificity_sensitivity(y_test, rf_pred)

# Print the specificity and sensitivity for each model
print(f"Logistic Regression: Specificity={specificity_logreg:.3f}, Sensitivity={sensitivity_logreg:.3f}")
print(f"k-NN: Specificity={specificity_knn:.3f}, Sensitivity={sensitivity_knn:.3f}")
print(f"Naive Bayes: Specificity={specificity_nb:.3f}, Sensitivity={sensitivity_nb:.3f}")
print(f"SVM: Specificity={specificity_svm:.3f}, Sensitivity={sensitivity_svm:.3f}")
print(f"Decision Tree: Specificity={specificity_dt:.3f}, Sensitivity={sensitivity_dt:.3f}")
print(f"Random Forest: Specificity={specificity_rf:.3f}, Sensitivity={sensitivity_rf:.3f}")
```

Logistic Regression has the highest specificity
KNN has the highest Sensitivity

K fold Validation


```python
#Importing cross validation score
from sklearn.model_selection import cross_val_score

k = 10 # number of folds
scores_knn = cross_val_score(knn, X, y, cv=k) 
scores_lr = cross_val_score(lr, X, y, cv=k)
scores_sv = cross_val_score(sv, X, y, cv=k)
scores_dt = cross_val_score(dt, X, y, cv=k)
scores_rf = cross_val_score(rf, X, y, cv=k)
scores_nb = cross_val_score(nb, X, y, cv=k)

#Storing scores of each classifier in scores list
scores = [scores_lr,scores_knn,scores_nb,scores_sv,scores_dt,scores_rf]

#Calculating average accuracies for k folds
average_score_logreg = scores_lr.mean()
average_score_knn = scores_knn.mean()
average_score_nb = scores_nb.mean()
average_score_svm = scores_sv.mean()
average_score_dt = scores_dt.mean()
average_score_rf = scores_rf.mean()

#Printing mean accuracies
print(f"Average performance of Logistic Regression is: {average_score_logreg*100:.2f}")
print(f"Average performance of K Nearest Neighbour is: {average_score_knn*100:.2f}")
print(f"Average performance of Naive-Bayes is: {average_score_nb*100:.2f}")
print(f"Average performance of SVM is: {average_score_svm*100:.2f}")
print(f"Average performance of Decision Tree is: {average_score_dt*100:.2f}")
print(f"Average performance of Random Forest is: {average_score_rf*100:.2f}")
```
